````python
from git import Repo
# we will also use Regex and time manipulation for this project, so lets add those as well
import time
import re as regex
# we might as well generate some plots
import matplotlib.pyplot as plt
````

We will need a repo to use. GitPython allows you to clone by using a simple function:

````python
# Repo.clone_from(repo_url, directory_to_clone_onto)
# This example will clone the repo 'lime-csharp' from 'takenet' onto the './lime-chsarp'
try:
    Repo.clone_from('https://github.com/takenet/lime-csharp', './lime-csharp')
except:
    print('Repo already cloned or folder not empty. Skipping')
````

If you already have a repo you want to use, you can simply select it:

````python
# Repo(repo_folder)
# We are selecting the above-cloned repo
repo = Repo('./lime-csharp')
````

We now have a cloned and referenced repo to use.

Lets get started then. How many files were edited on the last commit?

````python
allCommits = repo.iter_commits()
element = next(allCommits)
count = 0
for file in element.stats.files:
    count += 1
print('Number of files on latest commit: ', count)
print('Files: ', element.stats.files)
````

What about the first one?

````python
allCommits = repo.iter_commits()
element = next(allCommits)
count = 0
while True:
    try:
        element = next(allCommits)
    except StopIteration:
        break
for file in element.stats.files:
    count += 1
print('Number of files on first commit: ', count)
print('Files: ', element.stats.files)
````

Just to test some filters... what if we want to count only files from a certain type? Lets say we have a .NET project. We care mostly about the `.cs` files, so lets filter them:

````python
allCommits = repo.iter_commits()
element = next(allCommits)
count = 0
fileList = []
for file in element.stats.files:
    if file.endswith('.cs'):
        count += 1
        fileList.append(file)
print('Number of \'.cs\' files on first commit: ', count)
print('Files: ', fileList)
````

Lets make things more interesting now. What if we want to check the total number of lines of code of files on a single commit? Lets consider only `.cs` files again.

````python
import io

allCommits = repo.iter_commits()
commit = next(allCommits)
files = commit.stats.files
for file in files:
    if not file.endswith('.cs'):
        continue
    try:
        targetfile = commit.tree / file
    # files not found
    except KeyError:
        continue
    with io.BytesIO(targetfile.data_stream.read()) as f:
        num_lines = sum(1 for line in f)
        print(file, "\n\tLine Count: ", num_lines)
````

Lets get some interesting metrics now. What if we want to check the most *unstable* `.cs` file? In other words, which file is the most edited?

````python
import collections

allCommits = repo.iter_commits()
dictionary = {}
while True:
    try:
        commit = next(allCommits)
        for file in commit.stats.files:
            if not file.endswith('.cs'):
                continue
            # lets ignore Tests as well
            if 'Tests' in file:
                continue
            if file not in dictionary:
                dictionary[file] = 1
            else:
                dictionary[file] += 1
    except StopIteration:
        break
````

````python
count = 0
keyList = []
valueList = []
for key, value in sorted(dictionary.items(), key=lambda kv: kv[1], reverse=True):
    keyList.append(key)
    valueList.append(value)
    count += 1
    if count == 10:
        break

plt.bar(range(len(valueList)), valueList, align='center')
plt.xticks(range(len(keyList)), keyList, rotation='vertical')
````

And which are the most complex ones? One way to determine that is to check which files have the most lines.

Note: there are ways to improve that metric (ex: fan-in, fan-out). Here is shown only a simple example to motivate the reader.

````python
allCommits = repo.iter_commits()
dictionary = {}
while True:
    try:
        commit = next(allCommits)
        for file in commit.stats.files:
            if not file.endswith('.cs'):
                continue
            # lets ignore Tests as well
            if 'Tests' in file:
                continue
            if file not in dictionary:
                try:
                    targetfile = commit.tree / file
                # files not found
                except KeyError:
                    continue
                with io.BytesIO(targetfile.data_stream.read()) as f:
                    num_lines = sum(1 for line in f)
                dictionary[file] = num_lines
    except StopIteration:
        break
````

````python
count = 0
keyList = []
valueList = []
for key, value in sorted(dictionary.items(), key=lambda kv: kv[1], reverse=True):
    keyList.append(key)
    valueList.append(value)
    count += 1
    if count == 10:
        break

plt.bar(range(len(valueList)), valueList, align='center')
plt.xticks(range(len(keyList)), keyList, rotation='vertical')
````
