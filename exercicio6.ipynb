```python
from git import Repo
import time
from collections import OrderedDict
from operator import itemgetter

#get the porcentage of files inside fullAuthorList wich cover is less then minimum
def GetAbandonedPercentage(fullAuthorList, minThreshold=0.0, minimumFileCover=0.5):
    total = 0
    abandoned = 0
    for file in fullAuthorList:
        total += 1
        totalPoints = 0
        totalCovered = 0
        for author in fullAuthorList[file]:
            if((fullAuthorList[file])[author] >= 0 and (fullAuthorList[file])[author] > minThreshold):
                totalPoints += (fullAuthorList[file])[author]
                totalCovered += (fullAuthorList[file])[author]
            elif((fullAuthorList[file])[author] > minThreshold):
                totalPoints += ((fullAuthorList[file])[author] * -1)
        if(totalPoints > 0 and (totalCovered * 100)/totalPoints < minimumFileCover):
            abandoned += 1
    return (abandoned*100)/total


#returns the truck factor based on authoral list
def GetTruckFactor(fileAuthorsList, minThreshold=0.0, minimumAbandonedPercentageFiles=0.5, minimumFileCover=0.5):
    truckFactors = []
    fullAuthorList = {}
    for file in fileAuthorsList:
        for author in fileAuthorsList[file]:
            if author not in fullAuthorList:
                fullAuthorList[author] = 1
            else:
                fullAuthorList[author] += 1
    orderedFullAuthorList = OrderedDict(sorted(fullAuthorList.items(), key = itemgetter(1), reverse = True))
    for author in orderedFullAuthorList:
        truckFactors.append(author)
        for file in fileAuthorsList:
            for someAuthor in fileAuthorsList[file]:
                if(someAuthor == author and (fileAuthorsList[file])[someAuthor] >= 0 and (fileAuthorsList[file])[someAuthor] > minThreshold):
                    (fileAuthorsList[file])[someAuthor] *= -1
        if(GetAbandonedPercentage(fileAuthorsList, minimumFileCover) < minimumAbandonedPercentageFiles):
            return truckFactors
    return truckFactors                    
    

# prints the authoral list for each file with a given threshold as a minimum, returns the list as well
def GetFileAuthorsList(files, minThreshold=0.0):
    fileAuthorsList = {}
    for file in files:
        fileAuthors = {}
        mods = 0
        # count number of mods per author and absolute
        for author in files[file]:
            mods += 1
            if author not in fileAuthors:
                fileAuthors[author] = 1
            else:
                fileAuthors[author] += 1
        totalFileAuthors = len(fileAuthors)
        creator = list(fileAuthors.keys())[0]
        # Calculates author based on A(d,f) = C + 0.5*MD -0.1*MO
        for fileAuthor in fileAuthors:
            if fileAuthor == creator:
                fileAuthors[fileAuthor] = 1 + 0.5*fileAuthors[fileAuthor] - 0.1*(mods-fileAuthors[fileAuthor])
            else:
                fileAuthors[fileAuthor] = 0.5*fileAuthors[fileAuthor] - 0.1*(mods-fileAuthors[fileAuthor])
        # Normalizes it within [0,1] where 1 is the author with most 'points'
        for fileAuthor in fileAuthors:
            ordered = OrderedDict(sorted(fileAuthors.items(), key = itemgetter(1), reverse = True))
            owner = list(ordered.keys())[0]
            fileAuthors[fileAuthor] = fileAuthors[fileAuthor]/fileAuthors[owner]

        # Prints the calculated ownership for each file, excluding anyone below a certain threshold
        print(file)
        ordered = OrderedDict(sorted(fileAuthors.items(), key = itemgetter(1), reverse = True))
        for k, v in ordered.items():
            if v <= minThreshold:
                break
            print('\t', k, ':', v)
        print('\n')
        fileAuthorsList[file] = fileAuthors
    return fileAuthorsList
        
def GetFilesAndEditorsFromRepo(repo):
    commits = repo.iter_commits(reverse=True)
    files = {}
    for commit in commits: # From first to last
        for file in commit.stats.files:
            if file not in files:
                files[file] = []
                files[file].append(commit.author.name)
            else:
                files[file].append(commit.author.name)
    return files
```

**Calculando Truck-Factor :**
<br>
*(O cálculo é feito da seguinte forma: somando os pontos de cada arquivo, se obtem a pontuação total dele. Removemos o autor que mais tem commits no projeto, e então vemos quanto de pontuação cada arquivo do projeto perdeu. Caso exista mais que um mínimo definido de arquivos com mais de determinada porcentagem de pontuação perdida, esse autor é o único truck factor; caso contrário, removemos o segundo autor com mais commits, e assim vai até encontrar todos os truck-factors.*

```python
truckFactor = GetTruckFactor(authorList)
print(truckFactor)
```
