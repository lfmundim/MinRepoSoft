```python
from git import Repo
import time
from collections import OrderedDict
from operator import itemgetter

# prints the authoral list for each file with a given threshold as a minimum, returns the list as well
def GetFileAuthorsList(files, minThreshold=0.0):
    fileAuthorsList = []
    for file in files:
        fileAuthors = {}
        mods = 0
        # count number of mods per author and absolute
        for author in files[file]:
            mods += 1
            if author not in fileAuthors:
                fileAuthors[author] = 1
            else:
                fileAuthors[author] += 1
        totalFileAuthors = len(fileAuthors)
        creator = list(fileAuthors.keys())[0]
        # Calculates author based on A(d,f) = C + 0.5*MD -0.1*MO
        for fileAuthor in fileAuthors:
            if fileAuthor == creator:
                fileAuthors[fileAuthor] = 1 + 0.5*fileAuthors[fileAuthor] - 0.1*(mods-fileAuthors[fileAuthor])
            else:
                fileAuthors[fileAuthor] = 0.5*fileAuthors[fileAuthor] - 0.1*(mods-fileAuthors[fileAuthor])
        # Normalizes it within [0,1] where 1 is the author with most 'points'
        for fileAuthor in fileAuthors:
            ordered = OrderedDict(sorted(fileAuthors.items(), key = itemgetter(1), reverse = True))
            owner = list(ordered.keys())[0]
            fileAuthors[fileAuthor] = fileAuthors[fileAuthor]/fileAuthors[owner]

        # Prints the calculated ownership for each file, excluding anyone below a certain threshold
        print(file)
        ordered = OrderedDict(sorted(fileAuthors.items(), key = itemgetter(1), reverse = True))
        for k, v in ordered.items():
            if v <= minThreshold:
                break
            print('\t', k, ':', v)
        print('\n')
        fileAuthorsList.append(fileAuthors)
    return fileAuthorsList
        
def GetFilesAndEditorsFromRepo(repo):
    commits = repo.iter_commits(reverse=True)
    commitCount = 0
    files = {}
    for commit in commits: # From first to last
        commitCount += 1
        for file in commit.stats.files:
            if file not in files:
                files[file] = []
                files[file].append(commit.author.name)
            else:
                files[file].append(commit.author.name)
    return files
```
